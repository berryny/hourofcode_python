## Python Language Introduction

[Python](https://www.python.org/) is an interpreted, high-level and general-purpose programming language. Python's design philosophy emphasizes code readability with its notable use of significant whitespace. Its language constructs and object-oriented approach aim to help programmers write clear, logical code for small and large-scale projects.

Python was created in the late 1980s, and first released in 1991, by Guido van Rossum as a successor to the ABC programming language. Python 2.0, released in 2000, introduced new features, such as list comprehensions, and a garbage collection system with reference counting, and was discontinued with version 2.7 in 2020. Python 3.0, released in 2008, was a major revision of the language that is not completely backward-compatible and much Python 2 code does not run unmodified on Python 3.

## Python open() Function

The `open()` function opens a file, and returns it as a file object.
```
open(filename, mode)
```

Mode: `"r"` - Read - Default value. Opens a file for reading, error if the file does not exist

````python
# parse Department of Health and Mental Hygiene (DOHMH) CSV:
# parse CSV file

with open('coronavirus_data_by_boro.csv', 'r') as file:
    csvBoroData = csv.reader(file)
    for row in csvBoroData:
        print("**CSV COVID Boro Data**\n\t", row)
````

````python
# parse Department of Health and Mental Hygiene (DOHMH) CSV:
# parse CSV file

borough = input("Enter your borough: ")

with open('coronavirus_data_by_boro.csv', 'r') as file:
    csvMyBoroData = csv.reader(file)
    for num,boro_row in enumerate(csvMyBoroData):
#         print('num,boro_row\t',num,boro_row)
        if num > 0:
#             print(f"\nYou Entered: {borough}\n")
#             if borough == boro_row[0]:
            if borough.lower() == boro_row[0].lower():
                print("**FOUND IT**:\n", num,boro_row)
            else:
                print("Boro: NOT FOUND")
````

## Python JSON

JSON is a syntax for storing and exchanging data.

Python has a built-in package called `json`, which can be used to work with JSON data.

````python
# import the JSON 
import json
````

````python
# Template JSON information
fixed_json =  '{ "first_name":"John", "job":"Coder", "city":"New York"}'
fixed_json
````

````python
# Dynamic JSON information
dynamic_json =  '{ "first_name":"'+fname+'" , "job":"'+job+'", "city":"'+city+'"}'
dynamic_json
````

````python
# parse JSON:
parse_json = json.loads(dynamic_json)

# the result is a Python dictionary:
print(parse_json["job"])
````

````python
# parse Department of Health and Mental Hygiene (DOHMH) CSV:
# parse CSV file

with open('data_cityofnewyork.csv', 'r') as file:
    csvData = csv.DictReader(file)
    for row in csvData:
        print("dict\n",dict(row))
        data_timestamp = row['date_of_interest']
        convert_datastamp_format = datetime.strptime(data_timestamp,"%Y-%m-%dT%H:%M:%S.%f").strftime("%m/%d/%Y")
        
        print("Unformated dict data\n", convert_datastamp_format, row['case_count'], row['hospitalized_count'], row['death_count'])
#         print(f"Formated dict\n TimeStamp: {convert_datastamp_format} Case Count: {row['case_count']} Hospitalized Count: {row['hospitalized_count']} Death Count: {row['death_count']}")
````

````python
# Fake dataset
height = [boro_df_num[0][1], boro_df_num[1][1], boro_df_num[2][1]]
print('height',height)

bars = (boro_df_num[0][0], boro_df_num[1][0], boro_df_num[2][0])
print('bars',bars)

y_pos = np.arange(len(bars))

# Create bars and choose color
plt.bar(y_pos, height, color = (0.5,0.1,0.5,0.6))

# Add title and axis names
plt.title('By Boros')
plt.xlabel('BOROUGH_GROUP')
plt.ylabel('CASE_RATE')

# Limits for the Y axis
plt.ylim(0,6000)

# Create names
plt.xticks(y_pos, bars)

# Show graphic
plt.show()
````
